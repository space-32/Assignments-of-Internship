# **Clustering Algorithms Implementation**

# **1. K Means Clustering**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Range of cluster numbers to evaluate
range_n_clusters = list(range(2, 11))

# Store the silhouette scores for each k
silhouette_scores = []

# Loop through different numbers of clusters and calculate silhouette score
for k in range_n_clusters:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {k}, the average silhouette score is: {silhouette_avg}")

# Plot the Silhouette Scores
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title('Silhouette Score for Different Numbers of Clusters')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()

#Silhouette Score: Provides a quantitative measure that indicates how well the points are clustered (a score closer to 1 means better clustering)
```

# **2. Hierarchical Clustering**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Perform hierarchical/agglomerative clustering
Z = linkage(X, method='ward')  # 'ward' minimizes variance between clusters

# Plot the dendrogram
plt.figure(figsize=(10, 7))
plt.title('Dendrogram for Agglomerative Clustering')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
dendrogram(Z, truncate_mode='level', p=3)  # Show top 3 levels of the hierarchy
plt.show()

# Applying Agglomerative Clustering for k clusters (let's assume 3 clusters)
k = 3
agglomerative_clustering = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward')
cluster_labels = agglomerative_clustering.fit_predict(X)

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', marker='o')
plt.title(f'Agglomerative Clustering with {k} clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# **3.DBSCAN**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler

# Generate a synthetic dataset with 3 clusters and noise
X, y = make_blobs(n_samples=1000, centers=[[-2, -2], [2, 2], [5, 8]], cluster_std=[0.4, 0.5, 1.2], random_state=42)

# Standardize the dataset for better performance of DBSCAN
X = StandardScaler().fit_transform(X)

# Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], s=15, c='grey')
plt.title('Dataset with Noise and Varying Density')
plt.show()

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=10)
dbscan_labels = dbscan.fit_predict(X)

# Plot DBSCAN results
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='plasma', s=15)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Apply K-Means with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Plot K-Means results
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', s=15)
plt.title('K-Means Clustering (k=3)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# **Dimensionality Reduction Techniques**

# **Principal Component Analysis (PCA)**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the breast cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Standardize the dataset before applying PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce dimensionality
pca = PCA(n_components=2)  # Reducing to 2 components
X_pca = pca.fit_transform(X_scaled)

# Visualize the transformed data in 2D
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='coolwarm', edgecolor='k', s=50)
plt.title('PCA: Dimensionality Reduction to 2D')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Class')
plt.show()

# Apply PCA with all components to capture variance information
pca_full = PCA().fit(X_scaled)

# Plot the explained variance ratio for each component
plt.figure(figsize=(8, 6))
plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.show()

# Show the explained variance ratio for the first few components
print("Explained variance ratio for the first few components:")
print(pca_full.explained_variance_ratio_[:5])
```

# **t-SNE (t-Distributed Stochastic Neighbor Embedding)**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Load the breast cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Standardize the dataset before applying dimensionality reduction
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply t-SNE to reduce to 2 dimensions
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_tsne = tsne.fit_transform(X_scaled)

# Apply PCA for comparison
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plotting the t-SNE results
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='coolwarm', edgecolor='k', s=50)
plt.title('t-SNE: Dimensionality Reduction to 2D')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(label='Class')
plt.show()

# Plotting the PCA results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='coolwarm', edgecolor='k', s=50)
plt.title('PCA: Dimensionality Reduction to 2D')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Class')
plt.show()
```

# **Dimensionality Reduction Techniques Comparison: PCA vs t-SNE**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target
labels = data.target_names

# Standardize the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply t-SNE to reduce to 2 dimensions
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_tsne = tsne.fit_transform(X_scaled)

# Plotting the PCA results
plt.figure(figsize=(14, 6))

# PCA Visualization
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=80)
plt.title('PCA: Iris Dataset Reduced to 2D')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Class')

# t-SNE Visualization
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', edgecolor='k', s=80)
plt.title('t-SNE: Iris Dataset Reduced to 2D')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(label='Class')

plt.tight_layout()
plt.show()
```
